```python
import requests
import json

url = "https://data.api.abs.gov.au/rest/data/CPI/3.10001.10.50.Q?detail=dataonly&format=jsondata"

response = requests.get(url)
data_dict = response.json()

print("Data successfully loaded into a dictionary.")
```

```python
adat = data_dict["data"]["dataSets"][0]['series']['0:0:0:0:0']['observations']
```

```python
# [ {}, {}, {} ]
```

```python
uj = []
for kulcs in adat:
  # print(kulcs, float(adat[kulcs][0]))
  adatpont = {"index":kulcs, "adat":float(adat[kulcs][0])}
  # print(adatpont)
  uj.append(adatpont)
  # print(uj)
```

```python
open('uj.json','w').write(json.dumps(uj))
```
